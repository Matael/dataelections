Transformation adresse des BV -> Coords GPS
-------------------------------------------

On a un CSV un peu dégueu :

    "ID";"NOM";"CODCOMMUNE";"COMMUNE";"NUMERO";"VOIE";"ADRESSE";"THEME";"CATEGORIE";"ACTIVITES";"SECTEURS";"CANTON";"STATUT";"GESTIONNAIRE"
    8;Bureau de vote (Rest. scol. Claircigny-Boussinière);181;Le Mans;7;rue de Claircigny;7 rue de Claircigny;ADMINISTRATION;Service à la population;Bureaux de vote;BV : 88,89;CANTON LE MANS NORD OUEST;Public;LE MANS

On veut une liste : 

    18 Rue de l'Abbé Carton, 75014 Paris, France
    58 Rue Mirabeau, 49000 Angers, France

```python
import csv

tour = "premier_tour"

filename = "/home/matael/workspace/hackerspace/dataelections/data/BUREAUX_VOTE.csv"

with open(filename) as f:
    reader = csv.reader(f, delimiter=";")
    reader.next()
    
    crossref_bureaux_oldaddr = {}
    error_count = 0
    for l in reader:
        if l[3] == 'Le Mans':
            cp = 72000
        else: error_count += 1
        addr_str = "{}, {} {}, France".format(l[6],cp,l[3])
        crossref_bureaux_oldaddr[addr_str.decode('latin-1')] = [int(_) for _ in l[10].split(':')[1].split(',')]
        print(addr_str)
        
              
print("\n\nErrors : {}".format(error_count))
        
        
```

Conversion CSV Coords -> GeoJSON
--------------------------------

Ce site ( http://www.gpsfrance.net/liste-adresses-vers-coordonnees-gps ) nous a permis de retrouver des coordonnées GPS pour les BV, mainenant, on passe ça en GeoJSON.

```python
import csv
import json

filename = "/home/matael/workspace/hackerspace/dataelections/data/bureaux_vote_coords.csv"


geolist = []
with open(filename) as f:
    reader = csv.reader(f,delimiter=";")
    for l in reader:
        geolist.append(
            {
                "type": "Feature",
                "geometry": {
                    "type": "Point",
                    "coordinates": [l[1], l[0]]
                },
                "properties": {
                    "name": "Bureau de vote",
                    "secteurs": ','.join(map(str,crossref_bureaux_oldaddr[l[3].decode('utf-8')]))
                }
            }
        )

savefile = "/home/matael/workspace/hackerspace/dataelections/data/bureaux_vote_coords_{}.json".format(tour)

with open(savefile,'w') as f:
    f.write(json.dumps(geolist))
```

Récupération des données du scrutin
-----------------------------------

Tout est là : http://extra.lemans.fr/elections/premier_tour/

```python
from bs4 import BeautifulSoup
from requests import get

base_url = 'http://extra.lemans.fr/elections/{}/xml/'.format(tour)
```

### La liste des fichiers pour les bureaux

```python
get_liste = get('{}{}'.format(base_url,'Election.xml'))

bureaux_soup  = BeautifulSoup(get_liste.text).find_all('bureau')
```

```python
bureaux = {}
for b in bureaux_soup:
    num = int(b.find('numero').text)
    fichier = "{}.xml".format(b.find('fichier').text)
    candidats = BeautifulSoup(get("{}{}".format(base_url,fichier)).text).find_all('candidat')
    c_results = {}
    for c in candidats:
        c_results[int(c.find('intituler').text.split('-')[0])] = float(c.find('pourcentage').text.split('%')[0].replace(',','.'))
    bureaux[num] = c_results
        
```

Dessin des secteurs associés aux bureaux
========================================

```python
import csv
import re
import md5
import string

filename = "/home/matael/workspace/hackerspace/dataelections/data/VOIES_PAR_BUREAU_VOTE.csv"

folder = "/home/matael/workspace/hackerspace/dataelections/data/secteurs"

street_matcher = re.compile(r'^([^\(]+)\(([^\)]+)')
trans = string.maketrans('','')
noalph = trans.translate(trans,string.digits)


ref_dict = {}


for current in range(1,98):
    with open("{}/sect_{}_pre.csv".format(folder,current), 'w') as save:
        with open(filename) as f:
            reader = csv.reader(f, delimiter=";")
            reader.next()
    
            error_count = 0
            for l in reader:
                # extraction du nom de rue
            
                if int(l[6]) == current:
                    m = street_matcher.search(l[1])
                    street_name = "{} {}".format(m.groups()[1],m.groups()[0])
                    street_md5 = md5.new(street_name).digest()
        
                    nums = [l[_].translate(trans,noalph) for _ in range(3,7) if l[_]!='']
                    max_num = max(nums)
                    min_num = min(nums)
        
                    ref_dict[street_md5] = int(l[6])
        
                    save.write('{} {}, 72000 Le Mans, France\n'.format(min_num, street_name))
                    save.write('{} {}, 72000 Le Mans, France\n'.format(max_num, street_name))
              
    print("[Secteur {}] Errors : {}".format(current,error_count))
        
```

Exploitation des données sectorielles
=====================================

```python
import json

sect_filename = "data/secteurs.json"

# d'abord, on lit la liste des secteurs et on en fait un truc utilisable.
with open(sect_filename) as f:
    
    tmp = json.load(f)
    map_secteurs = {int(_['properties']['name'].split(' ')[1]):_['geometry']['coordinates'] for _  in tmp['features']}
    
```

```python
geolist = {"type": "FeatureCollection", "features": []}

for s in map_secteurs.keys():
    new_feature = {
        'type': 'Feature',
        'geometry': {
            'type': 'Polygon',
            'coordinates': map_secteurs[s]
        },
        'properties': {
            '_storage_options': {},
            'name': 'Secteur {}'.format(s)
        }
    }
    
    res = bureaux[s]
    
    # couleur du secteur
    res_sortedkeys = sorted(res, key=res.get, reverse=True)
    color = colors[tour][res_sortedkeys[0]]
    new_feature['properties']['_storage_options']['color'] = color
    
    # ajout du score de chacun
    for k,v in res.items():
        new_feature['properties'][listes[tour][k]] = "{} %".format(v)
        
    # ajout à la liste
    geolist['features'].append(new_feature)
```

```python
savefile = "/home/matael/workspace/hackerspace/dataelections/data/secteurs_results_{}.json".format(tour)

with open(savefile,'w') as f:
    f.write(json.dumps(geolist))
```
